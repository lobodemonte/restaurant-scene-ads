```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
# Setting path 
path = "./data/"

# Importing clustered Yelp data 
NYC_Yelp_clustered = pd.read_csv(path + "Yelp/NYC_Yelp_Clustered_k7.csv")
NYC_NTA_cluster_profile = pd.read_csv(path + "NYC_Yelp_NTA_Clustered_Profile_k2.csv")
```

```python
# Cleaning and dropping a column
NYC_Yelp_clustered.columns
NYC_Yelp_clustered = NYC_Yelp_clustered.drop(NYC_Yelp_clustered.columns[0],1)

# Factorizing cluster
stacked = NYC_Yelp_clustered[['Cluster']].stack()
NYC_Yelp_clustered[['Cluster']] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()
```

```python
NYC_Yelp_clustered = pd.DataFrame(NYC_Yelp_clustered)
NYC_Yelp_clustered.head()
```

```python
NYC_Yelp_clustered_melt = pd.melt(NYC_Yelp_clustered, 
                                  id_vars = ["Cluster"], 
                                  value_vars = ["ReviewCount", "Ratings", "Price"], 
                                 var_name = "Clusters")
plt.figure(figsize = (14, 8))
sns.boxplot(x = "Cluster", 
           y = "value",
           data = NYC_Yelp_clustered_melt,
           hue = "Clusters")

# dd=pd.melt(df,id_vars=['Group'],value_vars=['Apple','Orange'],var_name='fruits')
# sns.boxplot(x='Group',y='value',data=dd,hue='fruits')
```

```python
NYC_NTA_cluster_profile_melt = pd.melt(NYC_NTA_cluster_profile, 
                                  id_vars = ["NTA Cluster"], 
                                  value_vars = ["Yelp_Cluster1", "Yelp_Cluster2", "Yelp_Cluster3", "Yelp_Cluster4", "Yelp_Cluster5", "Yelp_Cluster6", "Yelp_Cluster7"], 
                                 var_name = "Clusters")
plt.figure(figsize = (12,10))
sns.boxplot(x = "NTA Cluster", 
           y = "value",
           data = NYC_NTA_cluster_profile_melt,
           hue = "Clusters")

# dd=pd.melt(df,id_vars=['Group'],value_vars=['Apple','Orange'],var_name='fruits')
# sns.boxplot(x='Group',y='value',data=dd,hue='fruits')
```
